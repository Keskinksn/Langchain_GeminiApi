```python
from IPython.display import display
from IPython.display import Markdown
import textwrap


def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

## Download The Document

Microsoft'tan bir ekip tarafından yürütülen, Fine Tuning ve RAG (Retrieval Augmented Generation) tekniklerinin bilgi tabanlı performanslarının karşılaştırıldığı bir çalışma.

Çalışmanın amacı büyük dil modellerinin (LLM) bilgi tabanı performansını artırmak amacıyla fine-tuning ve RAG (retrieval-augmented generation) yöntemlerinin etkinliklerini karşılaştırmaktır. LLM'lerin yeni bilgileri entegre etme ve mevcut bilgileri güncelleme becerilerini değerlendirerek her iki sürecin etkinliğini ortaya koyuyor.

📍Çalışma kapsamında; Mistral, Llama 2 ve Orca 2'nin 7b modelleri gibi gelişmiş LLM'ler kullanılmış ve embedding model olarak BGE-large-en embedding tercih edilmiş.

Performans dört ana yöntem üzerinden ölçülmüş:

1. LLM
2. LLM + Retrieval-Augmented Generation (RAG)
3. Fine-tune edilmiş LLM
4. ⁠Fine-tune edilmiş LLM + RAG

Sonuçlara bakıldığında benchmark sonuçlarının ikinci tekniğin (LLM+RAG) lehine olduğu gözlemlenmiş.

Nihayetinde QA (soru-cevap) sistemlerinde base modelin + RAG ile birleştirilmesinin diğer tekniklere nazaran çok daha başarılı olduğunu göstermiş.

```python
data_folder = p.cwd() / "data"
p(data_folder).mkdir(parents=True, exist_ok=True)

pdf_url = "https://arxiv.org/pdf/2312.05934.pdf"
pdf_file = str(p(data_folder, pdf_url.split("/")[-1]))

urllib.request.urlretrieve(pdf_url, pdf_file)
```

## Extract text from the PDF

```python
!pip install pypdf
```

```python
pdf_loader = PyPDFLoader(pdf_file)
pages = pdf_loader.load_and_split()
print(pages[5].page_content)
```

```python
context = "\n".join(str(p.page_content) for p in pages[:10])
print("The total words in the context: ", len(context))
```

```python
prompt_template = """Answer the question as precise as possible using the provided context. If the answer is
                    not contained in the context, say "answer not available in context" \n\n
                    Context: \n {context}?\n
                    Question: \n {question} \n
                    Answer:
                  """

prompt = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
```

```python
stuff_chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)
```

```python
question = "Fine-Tuning or Retrieval? Provide a detailed answer."


stuff_answer = stuff_chain(
    {"input_documents": pages[1:10], "question": question}, return_only_outputs=True
)
```

```python
pprint(stuff_answer)
```

```python
# Çalışma bulgularına göz önünde bulundurarak, fine-tuning ve retrieval-augmented
#generation (RAG) yöntemleri arasında bilgi enjeksiyonunda RAG'ın neden sürekli
#olarak daha iyi performans gösterdiğini anlamak için hangi faktörlere işaret ediyor?

question = "What are the findings obtained in the Experiments and Results?"

stuff_answer = stuff_chain(
    {"input_documents": pages[1:10], "question": question}, return_only_outputs=True
)

pprint(stuff_answer)
```

# RAG Pipeline: Embedding + LLM

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=0)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
```

```python
!pip install chromadb
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever()
```

```python
question = "What is a Knowledge Base Creation?"
docs = vector_index.get_relevant_documents(question)
```

```python
docs
```

```python
stuff_answer = stuff_chain(
    {"input_documents": docs, "question": question}, return_only_outputs=True
)
```

```python
pprint(stuff_answer)
```
